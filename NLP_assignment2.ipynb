```python
# converting value: string format
def combine_text(list_of_test):
    combined_text=''.join(list_of_test)
    return combined_text
```

```python
combined_data = {key:[combine_text(value)] for (key,value) in data.items()}
print(combined_data)
```

```python
import pandas as pd
data_frame = pd.DataFrame.from_dict(combined_data).transpose()
data_frame.columns = ['transcript']
data_frame = data_frame.sort_index()
data_frame
```

```python
import re
import string

def clean_text_round1(text):
    text = text.lower()
    text = re.sub('\[.*?\]','',text)
    text = re.sub('[%s]' %re.escape(string.punctuation),'',text)
    text = re.sub('\w*\d\w','',text)
    return text

round1 = lambda x:clean_text_round1(x)
```

```python
data_clean = pd.DataFrame(data_frame.transcript.apply(round1))
data_clean['transcript'][0]
```

```python
def clean_text_round2(text):
    '''Get rid of some additional punctuation and non-sensical text that was missed the first time around.'''
    text = re.sub('[‘’“”…]', '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\t+', ' ', text)
    return text

round2 = lambda x: clean_text_round2(x)
```

```python
data_clean = pd.DataFrame(data_clean.transcript.apply(round2))
data_clean.transcript[2]
```

```python
def clean_round3(text):
    text = re.sub(r"♪+", "", text)
    text = re.sub(r"  ", "", text)
    pattern = r"(\b\w+n)'t\b"  # Matches words ending with "n't"
    text = re.sub(pattern, r"\1 not", text)
    return text

round3 = lambda x:clean_round3(x)
```

```python
import joblib
data_clean = pd.DataFrame(data_clean.transcript.apply(round3))
data_clean.transcript[4]
```

```python
joblib.dump(data_clean, "mydata_clean.pkl")
```
